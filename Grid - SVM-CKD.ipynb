```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset= pd.read_csv('CKD.csv')
dataset
```

```python
# Ai cant red the word function so we will changes to numeric using #One Hot Encoding#
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
#to identify the value count of classified column(Dependent)
dataset["classification_yes"].value_counts()
```

```python
# to get the columns name we use below algorithm
dataset.columns
```

```python
#to sepearate independent and dependent values 
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes',]]
dep=dataset['classification_yes']
```

```python
#split the test & train test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep,dep,test_size =1/3, random_state = 0)

# sklearn standardscaler is used for fitting the best result
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#logistic regression model
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto','scale'],
             'C':[10,100,1000,2000,3000]}  



grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
grid.fit(X_train, y_train)

#predict the xtest 
y_pred = grid.predict(X_test)
#classifieds we cant get the rscore instead we using confusion matrix to get the avaregae score
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

```

```python
#predict the xtest 
y_pred = grid.predict(X_test)
#classifieds we cant get the rscore instead we using confusion matrix to get the avaregae score
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#once we get the classfied value, to analyse the report we have to do te classification report.
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_pred)
print(clf_report)
```
